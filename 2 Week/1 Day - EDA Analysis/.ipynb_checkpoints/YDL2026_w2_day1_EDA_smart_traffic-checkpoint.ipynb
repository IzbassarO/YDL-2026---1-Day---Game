## 2. Статанализ: среднее, медиана, std, квартили (задание №1)

Главный детектор лжи здесь — **разрыв между средним и медианой**. Если они расходятся, распределение
перекошено, и «среднее» уже вводит в заблуждение. Считаю и сразу смотрю на формальность.

```python
num_cols = ['traffic_volume', 'avg_vehicle_speed', 'vehicle_count_cars',
            'vehicle_count_trucks', 'vehicle_count_bikes', 'temperature', 'humidity']

summary = pd.DataFrame({
    'mean'   : df[num_cols].mean(),
    'median' : df[num_cols].median(),
    'std'    : df[num_cols].std(),
    'Q1'     : df[num_cols].quantile(0.25),
    'Q3'     : df[num_cols].quantile(0.75),
    'min'    : df[num_cols].min(),
    'max'    : df[num_cols].max(),
    'skew'   : df[num_cols].skew(),
})
# относительный разрыв среднее↔медиана: индикатор перекоса
summary['mean_vs_median_%'] = ((summary['mean'] - summary['median']).abs()
                               / summary['median'] * 100).round(2)
summary['CV_%'] = (summary['std'] / summary['mean'] * 100).round(1)  # разброс
summary.round(3)
```

**Вердикт.** Для каждой колонки `mean ≈ median` (разрыв < 2%), `skew ≈ 0`. Перекоса нет — распределения симметричны.

НО: разброс огромный (CV у `traffic_volume` ~50%, у счётчиков машин — за 50%). Симметрия + гигантский разброс +
ровные min/max (50–998, 20–80, 10–35…) — это почерк **равномерного распределения**, а не колокола.

⚠️ **Ловушка №1, которую ловим заранее:** `mean ≈ median` НЕ значит «нормальное распределение».
Это проверим формой в следующем разделе.

## 3. Форма распределения + нормальность (задание №4)

> *«Постройте распределение колонки: оно нормальное? есть два горба? где сидит выброс?»*

Гистограммы покажут форму, а тесты Шапиро–Уилка и Q-Q график вынесут формальный вердикт о нормальности.

```python
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
for ax, col in zip(axes.ravel(), num_cols):
    sns.histplot(df[col], kde=True, ax=ax, color='steelblue', edgecolor='white')
    ax.axvline(df[col].mean(),   color='red',   ls='--', lw=1.5, label='mean')
    ax.axvline(df[col].median(), color='green', ls=':',  lw=1.5, label='median')
    ax.set_title(col); ax.legend(fontsize=8)
axes.ravel()[-1].axis('off')
plt.suptitle('Распределения числовых признаков', fontsize=14, y=1.02)
plt.tight_layout(); plt.show()
```

```python
# Формальная проверка нормальности
# Shapiro–Uilk + D'Agostino. Если p < 0.05 -> отвергаем нормальность.
norm_rows = []
for col in num_cols:
    sample = df[col].sample(min(500, len(df)), random_state=42)  # Shapiro надёжен до ~5000
    w, p_sh = stats.shapiro(sample)
    k2, p_da = stats.normaltest(df[col])
    norm_rows.append({'feature': col,
                      'shapiro_p': p_sh, 'dagostino_p': p_da,
                      'нормальное?': 'да' if min(p_sh, p_da) > 0.05 else 'НЕТ'})
pd.DataFrame(norm_rows).set_index('feature').round(6)
```

```python
# Q-Q графики: точки на прямой => нормально. Отклонения по краям => нет.
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
for ax, col in zip(axes.ravel(), num_cols):
    stats.probplot(df[col], dist='norm', plot=ax)
    ax.set_title(f'Q-Q: {col}')
axes.ravel()[-1].axis('off')
plt.suptitle('Q-Q графики (норма = точки на прямой)', fontsize=14, y=1.02)
plt.tight_layout(); plt.show()
```

**Вердикт по форме.** Все тесты отвергают нормальность (p < 0.05). На гистограммах — плоские «столешницы»,
а не колокола; на Q-Q графиках точки ложатся буквой **S** (классика равномерного распределения: тяжёлые «плечи», нет хвостов).

📌 Подтвердилась улика из раздела 2: данные сгенерированы из **равномерного** распределения.
Один горб, без бимодальности. Вывод важен на будущее: тесты, которые требуют нормальности
(в т.ч. Pearson r и t-тест в их классическом виде), здесь надо применять с осторожностью.

## 4. Box plot + выбросы: IQR и z-score (задание №5)

> *«Стандартизуйте колонку и вытащите точки, где |z| больше 2 или 3. Это кандидаты в аномалии.»*

Box plot показывает медиану, квартили и усы (точки за усами — кандидаты в выбросы по правилу 1.5·IQR).
Параллельно считаю z-score.

```python
# Box plot на стандартизованных данных, чтобы признаки с разным масштабом были на одной картинке
z = (df[num_cols] - df[num_cols].mean()) / df[num_cols].std()
plt.figure(figsize=(14, 6))
sns.boxplot(data=z, orient='h', palette='Set2')
plt.axvline(0, color='gray', lw=1)
plt.title('Box plot (z-стандартизованные признаки)')
plt.xlabel('z-score'); plt.tight_layout(); plt.show()
```

```python
# Считаем выбросы двумя методами
rows = []
for col in num_cols:
    s = df[col]
    Q1, Q3 = s.quantile([0.25, 0.75]); IQR = Q3 - Q1
    iqr_out = ((s < Q1 - 1.5*IQR) | (s > Q3 + 1.5*IQR)).sum()
    zscore = (s - s.mean()) / s.std()
    rows.append({'feature': col,
                 'выбросов_IQR': int(iqr_out),
                 '|z|>2': int((zscore.abs() > 2).sum()),
                 '|z|>3': int((zscore.abs() > 3).sum())})
pd.DataFrame(rows).set_index('feature')
```

**Вердикт по выбросам.** По IQR — **ноль** выбросов почти везде; по `|z|>3` — тоже ноль.

Это прямое следствие равномерного распределения: у него нет хвостов, значит и «аномалиям» взяться неоткуда.
В *реальных* городских данных пустой список выбросов был бы тревожным сигналом
(«а датчики вообще работают?»). Здесь же это финальное подтверждение: **данные искусственные**.

⚠️ **Ловушка №2:** отсутствие выбросов ≠ «чистые качественные данные». Иногда это значит, что данные ненастоящие.

## 5. Корреляции между признаками (задание №2)

Считаю матрицу Пирсона и сразу напоминаю себе: **корреляция — это ещё не причинность**.
Сначала смотрю, есть ли вообще что-то выше уровня шума.

```python
corr = df[num_cols].corr()
plt.figure(figsize=(9, 7))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            vmin=-1, vmax=1, square=True, linewidths=.5)
plt.title('Матрица корреляций (Pearson r)')
plt.tight_layout(); plt.show()
```

```python
# Самая сильная по модулю пара признаков + её p-value
pairs = []
for i, a in enumerate(num_cols):
    for b in num_cols[i+1:]:
        r, p = stats.pearsonr(df[a], df[b])
        pairs.append({'пара': f'{a} ~ {b}', 'r': r, 'p_value': p, '|r|': abs(r)})
pairs = pd.DataFrame(pairs).sort_values('|r|', ascending=False)
print('Самые сильные связи (по |r|):')
pairs.head(5).round(4)
```

**Вердикт по корреляциям.** Максимальное |r| во всём датасете ≈ 0.05.
Это **статистический ноль** — признаки взаимно независимы.

Связи нет ни настоящей, ни даже случайно-заметной на уровне всей выборки.
Что отлично подводит нас к главному фокусу лабы: если связи нет, то откуда берутся «значимые» результаты,
которые мы сейчас будем добывать? Это и есть призрак.

## 6. Pair plot и *действительность* корреляции: confounding (ключевой раздел)

> Проверяю не «есть ли цифра корреляции», а **реальна ли связь**. Цифра может появиться из-за
> **confounding** — третьей переменной, которая двигает оба признака сразу, создавая иллюзию прямой связи.
> Инструмент проверки — **частная корреляция** (partial correlation): держим третий признак постоянным и смотрим, выживет ли связь.

```python
# Pair plot по подвыборке (2000 точек на 7 признаков -> берём 600 для скорости отрисовки)
sample = df.sample(600, random_state=42)
g = sns.pairplot(sample[num_cols], diag_kind='kde', plot_kws=dict(s=12, alpha=0.4))
g.fig.suptitle('Pair plot: ищем структуру глазами', y=1.01)
plt.show()
```

На pair plot — **бесформенные облака**. Ни линий, ни кластеров, ни изгибов. Глаз не находит связи —
это согласуется с матрицей корреляций. Но проверим формально механизм confounding на самой «сильной» паре.

```python
def partial_corr(d, x, y, z):
    # Частная корреляция x и y при контроле z.
    # Формула: (r_xy - r_xz*r_yz) / sqrt((1-r_xz^2)(1-r_yz^2)).
    r_xy = d[x].corr(d[y]); r_xz = d[x].corr(d[z]); r_yz = d[y].corr(d[z])
    return (r_xy - r_xz*r_yz) / np.sqrt((1 - r_xz**2) * (1 - r_yz**2))

# Берём топ-пару из раздела 5 и контролируем все остальные признаки по очереди
top = pairs.iloc[0]['пара'].split(' ~ ')
x, y = top
print(f'Проверяемая пара: {x} ~ {y}')
print(f'Простая корреляция r_xy = {df[x].corr(df[y]):.4f}\n')
print('Частная корреляция при контроле третьей переменной:')
for z_var in [c for c in num_cols if c not in (x, y)]:
    print(f'  держим постоянным {z_var:22s}: r = {partial_corr(df, x, y, z_var):+.4f}')
```

**Что показывает частная корреляция.** Даже исходные ~0.05 не «очищаются» во что-то осмысленное —
связь как была шумом, так и осталась. Это и есть способ ловить confounding: *если бы* связь держалась на
скрытом факторе, при его фиксации она бы заметно поменялась. Здесь меняться нечему — связи нет.

Теперь покажу, **как confounding выглядит, когда он реален** — на честно сконструированном примере,
чтобы инструмент был понятен:

```python
# УЧЕБНАЯ ДЕМОНСТРАЦИЯ confounding (данные строим сами, помечаем явно).
# Сценарий: жара Z поднимает И продажи мороженого X, И число тепловых ударов Y.
# Прямой связи X->Y нет, но корреляция X~Y будет высокой -> ложный вывод.
n = 1000
Z = RNG.uniform(15, 40, n)               # температура (скрытый конфаундер)
X = 2*Z + RNG.normal(0, 5, n)            # продажи мороженого
Y = 1.5*Z + RNG.normal(0, 5, n)          # тепловые удары
demo = pd.DataFrame({'icecream': X, 'heatstroke': Y, 'temp': Z})

r_naive = demo['icecream'].corr(demo['heatstroke'])
r_part  = partial_corr(demo, 'icecream', 'heatstroke', 'temp')
print(f'Наивная корреляция icecream ~ heatstroke : r = {r_naive:.3f}  (выглядит как сильная связь!)')
print(f'Частная при контроле температуры          : r = {r_part:.3f}  (связь испарилась)')

fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(data=demo, x='icecream', y='heatstroke', hue='temp',
                palette='viridis', s=20, ax=ax[0])
ax[0].set_title(f'Кажется: мороженое -> тепловой удар (r={r_naive:.2f})')
# раскрашиваем по температуре: видно, что точки выстраиваются именно вдоль неё
sns.scatterplot(data=demo, x='temp', y='heatstroke', s=20, color='tomato', ax=ax[1])
ax[1].set_title('Правда: оба растут от температуры')
plt.tight_layout(); plt.show()
```

**Вердикт по confounding.** Слева связь `icecream ~ heatstroke` выглядит убедительной (r ≈ 0.9),
но это иллюзия: при контроле температуры она падает почти до нуля. Мороженое не вызывает тепловой удар —
оба признака тянет за собой жара.

📌 **Главный навык раздела:** прежде чем поверить в корреляцию из pair plot, спрашивай *«а нет ли третьего
фактора?»* и проверяй частной корреляцией. В нашем реальном датасете проверять, к счастью, нечего — связи и так нет.

## 7. Сравнение двух групп + t-тест (задание №8)

> *«Прогоните t-тест, получите p-value и усомнитесь: эффект правда большой или просто значимый?»*

Сравниваю среднюю скорость там, где была авария, и там, где не было. Считаю p-value И **размер эффекта** (Cohen's d) —
потому что на больших выборках «значимым» становится даже микроскопическое различие.

```python
g0 = df.loc[df['accident_reported'] == 0, 'avg_vehicle_speed']
g1 = df.loc[df['accident_reported'] == 1, 'avg_vehicle_speed']

t, p = stats.ttest_ind(g0, g1, equal_var=False)  # Welch
# Cohen's d — размер эффекта
pooled = np.sqrt(((g0.std()**2) + (g1.std()**2)) / 2)
d = (g1.mean() - g0.mean()) / pooled

print(f'Нет аварии (n={len(g0)}): mean speed = {g0.mean():.2f}')
print(f'Есть авария (n={len(g1)}): mean speed = {g1.mean():.2f}')
print(f'Разница средних : {g1.mean() - g0.mean():+.2f} км/ч')
print(f't = {t:.3f},  p-value = {p:.3f}')
print(f"Cohen's d = {d:.3f}  (0.2 — малый, 0.5 — средний, 0.8 — большой эффект)")

plt.figure(figsize=(8, 5))
sns.boxplot(x='accident_reported', y='avg_vehicle_speed', data=df, palette='Set3')
plt.title('Скорость: авария vs нет аварии'); plt.tight_layout(); plt.show()
```

**Вердикт по t-тесту.** Разница средних — доли км/ч, `Cohen's d` крошечный (|d| << 0.2),
p-value большой → различия нет. Здесь честно: тест говорит «не значимо».

Но запомни механику: будь выборка в 50 раз больше, та же ничтожная разница дала бы p < 0.05.
**p-value отвечает на вопрос «есть ли разница вообще», а не «велика ли она».** Размер эффекта — обязателен.

## 8. 👻 Охота на призрак: p-hacking (Особый вызов)

> *«Переберите пары и найдите впечатляющую корреляцию или значимое различие, которое на самом деле чистый шум.
> Покажите ловушку: вот цифра, которая убеждает, и вот почему она ничего не значит.»*

Мы знаем, что данные — шум. Сейчас я нарежу из них «открытия». Принцип p-hacking прост:
**проверь достаточно много гипотез — и случайность сама подарит тебе «значимый» результат.**

```python
# Перебираем МНОГО гипотез: корреляция каждой числовой пары внутри каждой категории погоды.
# При истинном отсутствии связи ~5% тестов всё равно дадут p < 0.05 — просто по случайности.
hypotheses = []
for w in df['weather_condition'].unique():
    sub = df[df['weather_condition'] == w]
    for i, a in enumerate(num_cols):
        for b in num_cols[i+1:]:
            r, p = stats.pearsonr(sub[a], sub[b])
            hypotheses.append({'погода': w, 'пара': f'{a} ~ {b}',
                               'n': len(sub), 'r': r, 'p_value': p})
H = pd.DataFrame(hypotheses)
n_tests = len(H)
n_sig = (H['p_value'] < 0.05).sum()
print(f'Проверено гипотез: {n_tests}')
print(f'«Значимых» (p<0.05): {n_sig}  ({n_sig/n_tests*100:.1f}% — почти ровно те самые 5% случайности)')
print(f'\nОжидалось ложных при чистом шуме: ~{0.05*n_tests:.0f}')
```

```python
# Достаём самый «убедительный» призрак и рисуем его с линией регрессии и R^2
ghost = H.loc[H['p_value'].idxmin()]
a, b = ghost['пара'].split(' ~ ')
sub = df[df['weather_condition'] == ghost['погода']]
slope, intercept, r_val, p_val, se = stats.linregress(sub[a], sub[b])

print('НАШ ПРИЗРАК (самый маленький p-value из всех):')
print(f"  погода = {ghost['погода']},  {a} ~ {b}")
print(f"  r = {r_val:.3f},  R^2 = {r_val**2:.3f},  p = {p_val:.4f},  n = {len(sub)}")

plt.figure(figsize=(8, 5))
sns.regplot(x=a, y=b, data=sub, scatter_kws=dict(s=20, alpha=0.5),
            line_kws=dict(color='red'))
plt.title(f"«Открытие»: при погоде '{ghost['погода']}' {a} влияет на {b}?\n"
          f"r={r_val:.2f}, p={p_val:.3f} — выглядит значимо!")
plt.tight_layout(); plt.show()
```

```python
# Поправка на множественные сравнения (Bonferroni): порог делим на число тестов.
alpha = 0.05
bonf = alpha / n_tests
n_survive = (H['p_value'] < bonf).sum()
print(f'Порог Bonferroni: {bonf:.6f}')
print(f'Пережили честную поправку: {n_survive} из {n_tests}')
print('\nВывод: ни один «призрак» не выживает после поправки на число попыток.')
```

**Вердикт по охоте на призрак.** Перебрав ~50 гипотез на заведомо случайных данных, мы «нашли» несколько
связей с p < 0.05 — ровно те ~5%, что обещает случайность. Самая красивая из них имеет приличный r и убедительный
график с линией регрессии. **И она не значит ничего:** после поправки Bonferroni не выживает ни одна.

📌 Это и есть статистическое ядро того, почему модели переобучаются, а R²=0.99 умеет врать:
если перебрать достаточно вариантов, шум всегда подсунет «закономерность».
Самое ценное открытие дня — не то, что данные сказали, а то, где они чуть нас не обманули.

## 9. Проверка руками (задание №3, привычка №3)

> *«Любой вывод агента сверяйте руками на маленьком куске. Запустилось — это ещё не значит, что правда.»*

Беру первые 10 строк и пересчитываю среднюю температуру вручную, без `.mean()`.

```python
head10 = df['temperature'].head(10).tolist()
manual_mean = sum(head10) / len(head10)          # руками
pandas_mean = df['temperature'].head(10).mean()  # машиной
print('Первые 10 значений temperature:', [round(v, 2) for v in head10])
print(f'Сумма / 10 (руками) : {manual_mean:.6f}')
print(f'pandas .mean()      : {pandas_mean:.6f}')
print('Совпадает:', np.isclose(manual_mean, pandas_mean))
```

**Вердикт.** Ручной счёт и `pandas` сошлись до шестого знака. Машине здесь можно верить —
но проверили это мы сами, а не поверили на слово.
